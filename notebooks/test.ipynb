```python
import pandas as pd
import numpy as np

path_pattern_cm = "D:/Projects/PyData_Assessment/data/crick_df.csv"
paths = "D:/Projects/PyData_Assessment/data/WorldCup_Stats/1975_Match_Stats.csv"

df = pd.read_csv(path_pattern_cm)
```
